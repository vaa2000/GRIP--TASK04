# Finding Correlation Matrix

```python
# get correlation matrix
corr = df.corr()
corr
#fig, ax = plt.subplots()
```

The correlation coefficient ranges in value between -1.0 and +1.0. To interpret the correlation coefficient, we must consider both its sign (positive or negative) and its absolute value. A perfect positive correlation has a coefficient of 1.0; a perfect negative correlation has a coefficient of -1.0. When there is no association between two variables, the correlation coefficient has a value of 0.

# Year-wise count of Terrorist Attacks

```python
plt.figure(figsize=(100,100))
unique_year = df['Year'].unique()
year_wise_count = df['Year'].value_counts()
plt.plot(unique_year, year_wise_count[:])
plt.xlabel("Year")
plt.xticks(rotation=90)
plt.ylabel("Count of Terrorist attacks in that particular Year")
plt.title("Count of Terrorist Attacks Year-wise")
plt.show()
```

```python

killed = df.loc[df['Killed'] == 1].groupby(df['Year']).count()
wounded = df.loc[df['Wounded'] == 1].groupby(df['Year']).count()
plt.plot(unique_year,killed,color='red',label="Killed",marker='o')#1),#marker='o' shows point sthrogh which line is passing
plt.plot(unique_year,wounded,color='green',label="Wounded",linestyle="dashed",marker='*')#2)
plt.xlabel("Time")#to give some title on x-axis
plt.ylabel("Count")

plt.title("Counts over Time")#to show some title to our graph

#lets say red line denoteds price of apple and green line that of kiwi
#so give labels like changes in 1) and 2)
#plt.legend()
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

#if u are interested to get dotted line u can also try that

plt.show()
```

# Country-wise count of Terrorist Attacks

```python
plt.figure(figsize=(100,100))
unique_country = df['Country'].unique()
country_wise_count = df['Country'].value_counts()
plt.plot(unique_country, country_wise_count[:])
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Count of Terrorist attacks in that particular Country")
plt.title("Count of Terrorist Attacks Country-wise")
plt.show()
```
